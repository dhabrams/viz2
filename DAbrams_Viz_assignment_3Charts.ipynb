## All Charts Reflect Number of Performances of the Grateful Dead 1965 - 1995
(data may not be complete)

```python
%matplotlib inline
import matplotlib
import seaborn as sns
matplotlib.rcParams['savefig.dpi'] = 144
import pandas as pd
from bokeh.charts import Histogram
from bokeh.plotting import show
from bokeh.io import output_notebook
from bokeh.charts import defaults
from bokeh.charts import Bar, output_file, show
defaults.width = 700
defaults.height = 400
output_notebook()
```

```python
Months_WeekdayDF = pd.read_excel('GD_shows_by_Month_weekday.xlsx')
pvalue='Month'
p1 = Bar(Months_WeekdayDF, pvalue, values='Kounter', title="Shows by Day of Week by Month",legend=True,group='Weekday')
p1.yaxis.axis_label = "Number of Shows"
p1.legend.orientation = "horizontal"
defaults.width = 1200
defaults.height = 500
show(p1)
```

```python
States_gt50 = pd.read_excel('States_gt50.xlsx') 
defaults.width = 800
defaults.height = 500
VBar =Bar(States_gt50, 'State',values='Number_of_Shows',legend=False,group='State',title="States with 50 or more Shows")
VBar.legend.orientation = "horizontal"
VBar.legend.location = "top_center"
VBar.xaxis.axis_label =  "State"
VBar.yaxis.axis_label =  "Number of Shows"
show(VBar)
```

```python
hist = Histogram(Months_WeekdayDF['Year'],title="#Shows by Years")
hist.yaxis.axis_label =  "Number of Shows"
show(hist)
```
